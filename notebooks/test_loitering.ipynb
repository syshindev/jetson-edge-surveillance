```python
import cv2

cap = cv2.VideoCapture("C:/Users/gmission/Desktop/video-test/GX012762.MP4")
ret, frame = cap.read()
cap.release()

points = []

def click(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        points.append([x, y])
        cv2.circle(frame, (x, y), 5, (0, 0, 255), -1)
        if len(points) > 1:
            cv2.line(frame, tuple(points[-2]), tuple(points[-1]), (0, 255, 0), 2)
        cv2.imshow("Draw Zone - Press Q when done", frame)

cv2.imshow("Draw Zone - Press Q when done", frame)
cv2.setMouseCallback("Draw Zone - Press Q when done", click)
cv2.waitKey(0)
cv2.destroyAllWindows()

print(f"Zone polygon: {points}")
```

```python
import sys
sys.path.append("../ai/src")
import cv2
from ultralytics import YOLO
from events.intrusion import Zone
from events.loitering import LoiteringDetector

model = YOLO("yolo11n.pt")
zone = Zone("test_area", [[839, 601], [1212, 582], [1244, 719], [859, 756]])
loiter = LoiteringDetector([zone], threshold=3.0)

cap = cv2.VideoCapture("C:/Users/gmission/Desktop/video-test/GX012762.MP4")

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(frame, persist=True, conf=0.5, classes=[0], verbose=False)
    events = loiter.check(results)

    if events:
        for e in events:
            print(f"Frame {frame_count}: Loitering ID {e['track_id']} in {e['zone']} ({e['duration']}s)")

    frame_count += 1
    if frame_count > 1000:
        break

cap.release()
print("Done")
```
