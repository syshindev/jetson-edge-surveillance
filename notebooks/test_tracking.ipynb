# Tracking Sanity Check

Verify that **YOLO11n + BoT-SORT** (Ultralytics' default tracker) assigns stable track IDs on top of detection boxes.

**Why tracking matters for DOBER:**
Event logic (line crossing, intrusion duration, loitering) needs to follow *the same person* across frames. A plain detector has no cross-frame identity — we need a tracker that links boxes over time.

**Why BoT-SORT:**
- Bundled with Ultralytics (no extra install)
- Uses ReID + motion (Kalman) — robust to brief occlusions
- Default choice in prior KISA surveillance work

**Test image:** `bus.jpg` (single frame — we just confirm `track_id` is populated; multi-frame stability is validated later on real video).

## 1. Run tracker

`model.track(..., persist=True)` tells the tracker to keep state across calls — essential for video, optional on a single frame but good to keep on so this cell is copy-paste friendly.

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
results = model.track(source="bus.jpg", persist=True)

for box in results[0].boxes:
    name = results[0].names[int(box.cls[0])]
    conf = float(box.conf[0])
    track_id = int(box.id[0]) if box.id is not None else None
    bbox = box.xyxy[0].tolist()
    print(f"ID:{track_id} {name}: {conf:.2f}, bbox: {bbox}")
```

## 2. Results

| Track ID | Class  | Confidence |
|----------|--------|------------|
| 1        | bus    | 0.94       |
| 2        | person | 0.89       |
| 3        | person | 0.88       |
| 4        | person | 0.86       |
| 5        | person | 0.62       |

- 5 tracks assigned on a single frame → ID population confirmed.
- Bounding boxes match the detection-only results from `test_detection.ipynb`, so the tracker is not perturbing detection quality.
